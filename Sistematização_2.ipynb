```python
%%latex
Código efetuado para plotar os gráficos da transformada de Fourier, em domínio de tempo e frequência, sobre a função $f(t)\\$
Quanto a função fiquei em dúvida se houve algum erro de digitação na confecção do docs da Sistematização 2, então para não correr riscos plotei para ambas as possíveis funções, sendo elas:
$$f(t) = e^{-|t|}*cos(10t)\\ ou \\g(t) = (e-|t|)*cos(10t)$$
```

```python
# Função que define f(t)
def f(t):
    return (np.exp(- np.abs(t)) ) * np.cos(10 * t)

def g(t):
    return (np.exp(1) - np.abs(t) ) * np.cos(10 * t)

# Função para plotar o sinal e sua transformada de Fourier
def plot_signal_and_fourier(duration, points, function):
    t = np.linspace(-duration / 2, duration / 2, points, endpoint=False)
    signal = function(t)

    # Calculando a transformada de Fourier
    fft_result = np.fft.fft(signal, points)
    freq = np.fft.fftfreq(points, (t[1] - t[0]))

    # Plotando o sinal no domínio do tempo
    plt.figure(figsize=(14, 10))

    plt.subplot(2, 2, 1)
    plt.plot(t, signal)
    if function == f:
        plt.title(f'Sinal f(t) no Domínio do Tempo (N = {points})')
    else:
        plt.title(f'Sinal g(t) no Domínio do Tempo (N = {points})')
    plt.xlabel('Tempo (s)')
    plt.ylabel('Amplitude')
    plt.grid()
    plt.xlim(np.min(t), np.max(t))  # Limitando o eixo x para melhor visualização

    # Plotando o sinal no domínio da frequencia
    plt.subplot(2, 2, 2)
    plt.plot(freq, np.abs(fft_result))
    if function == f:
        plt.title(f'Sinal f(t) no Domínio da Frequência (N = {points})')
    else:
        plt.title(f'Sinal g(t) no Domínio da Frequência (N = {points})')
    plt.xlabel('Frequência (Hz)')
    plt.ylabel('Magnitude')
    plt.grid()
    plt.ylim(bottom=0) # Limitando o eixo y para remover valores indesejados gerados pela função fft do numpy
    plt.xlim(np.min(freq), np.max(freq))  # Limitando o eixo x para melhor visualização

    plt.tight_layout()
    plt.show()

# Parâmetros
sample_rates = [256, 512, 1024, 2048]  # Taxas de amostragem
duration = 10.0                        # Duração em segundos

# Plotando o sinal e sua transformada de Fourier para diferentes números de pontos
for points in sample_rates:
    plot_signal_and_fourier(duration, points, f)


for points in sample_rates:
    plot_signal_and_fourier(duration, points, g)

```
